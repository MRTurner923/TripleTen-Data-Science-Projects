## Aggregate data per user

````python
# Calculate the number of calls made by each user per month. Save the result.
calls['month'] = pd.to_datetime(calls['call_date']).dt.month
calls_per_user = calls.groupby(['user_id', 'month']).size().reset_index(name = 'total_calls')
calls_per_user
````

````python
# Calculate the amount of minutes spent by each user per month. Save the result.
minutes_per_user = calls.groupby(['user_id', 'month'])['duration'].sum().reset_index(name = 'total_minutes')
minutes_per_user
````

````python
# Calculate the number of messages sent by each user per month. Save the result.
messages['month'] = pd.to_datetime(messages['message_date']).dt.month
messages_per_user = messages.groupby(['user_id', 'month']).size().reset_index(name = 'total_messages')
messages_per_user
````

````python
# Calculate the volume of internet traffic used by each user per month. Save the result.
internet['month'] = pd.to_datetime(internet['session_date']).dt.month
internet_per_user = internet.groupby(['user_id', 'month'])['mb_used'].sum().reset_index()
internet_per_user['mb_used'] = np.ceil(internet_per_user['mb_used'])
internet_per_user
````

````python
# Merge the data for calls, minutes, messages, internet based on user_id and month
agg_calls = calls.groupby(['user_id', 'month']).agg({'duration':'sum'}).reset_index()
agg_messages = messages.groupby(['user_id', 'month']).size().reset_index(name = 'messages')
agg_internet = internet.groupby(['user_id', 'month']).agg({'mb_used':'sum'}).reset_index()

df_merged = agg_calls.merge(agg_messages, on = ['user_id', 'month'], how = 'outer')
df_merged = df_merged.merge(agg_internet, on = ['user_id', 'month'], how = 'outer')
df_merged
````

````python
# Add the plan information
df_merged = df_merged.merge(users[['user_id', 'plan']], on = 'user_id', how = 'left')
df_full = df_merged.merge(plans, left_on = 'plan', right_on = 'plan_name', how = 'left')
df_full_filled = df_full.fillna(0)
df_full_filled
````

<div class="alert alert-success" style="border-radius: 15px; box-shadow: 4px 4px 4px; border: 1px solid ">
<h2> Reviewer's comment </h2>
    
Correct! We indeed need to use the `outer` merging, since we don't want to lose those clients we made some calls, but who did not use internet and/or send messages, for instance.
</div>

````python
# Calculate the monthly revenue for each user
def calculate_revenue(row):
    plan_row = plans[plans['plan_name'] == row['plan']]
    
    minutes_charge = 0
    messages_charge = 0
    internet_charge = 0

    if row['duration'] > plan_row['minutes_included'].values[0]:
        minutes_charge = (row['duration'] - plan_row['minutes_included'].values[0]) * plan_row['usd_per_minute'].values[0]
    
    if row['messages'] > plan_row['messages_included'].values[0]:
        messages_charge = (row['messages'] - plan_row['messages_included'].values[0]) * plan_row['usd_per_message'].values[0]

    excess_mb = row['mb_used'] - plan_row['mb_per_month_included'].values[0]
    if excess_mb > 0:
        internet_charge = np.ceil((excess_mb / 1024)) * plan_row['usd_per_gb'].values[0]
    
    additional_charges = minutes_charge + messages_charge + internet_charge
    
    total_revenue = additional_charges + plan_row['usd_monthly_pay'].values[0]
    return total_revenue

df_full_filled['monthly_revenue'] = df_full_filled.apply(calculate_revenue, axis=1)

print(df_full_filled[['user_id', 'month', 'monthly_revenue']])
````

## Test statistical hypotheses

The null hypothesis for this test would be that the average revenue from users of the Ultimate and Surf plans does not differ
The althernative hypothesis is that the average revenue from users of the Ultimate and Surf plan does differ.
For this test, I'll be choosing my alpha to be 0.05, giving me a 95% confidence level 

<div class="alert alert-success" style="border-radius: 15px; box-shadow: 4px 4px 4px; border: 1px solid ">
<h2> Reviewer's comment </h2>

    
Correct. The null hypothesis always includes the equality sign. This is because the test does not understand exactly how we set the problem: we can say, for instance, let's make sure that they are not equal. Or that they are equal. And regardless of the purpose of our study, we always put equality at null hypothesis. Then the test result is interpreted correctly.

</div>   

````python
surf_plan_rev = df_full_filled[df_full_filled['plan'] == 'surf']['monthly_revenue']
ultimate_plan_rev = df_full_filled[df_full_filled['plan'] == 'ultimate']['monthly_revenue']
alpha = 0.05
results = st.ttest_ind(surf_plan_rev, ultimate_plan_rev)
print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print('We reject the null hypothesis')
else:
    print('We cannot reject the null hypothesis')
````

The null hypothesis for this test would be that the average revenue from users in the NY-NJ region does not differ from other regions. The althernative hypothesis is that the average revenue from users in the NY-NJ area does differ from other regions. For this test, I'll be choosing my alpha to be 0.05, giving me a 95% confidence level in the results of my analysis.

````python
df_full_filled_city = df_full_filled.merge(users[['user_id', 'city']], on = 'user_id', how = 'left')
avg_rev_ny_nj = df_full_filled_city[df_full_filled_city['city'].str.contains('NY-NJ')]['monthly_revenue']
avg_rev_other_regions = df_full_filled_city[df_full_filled_city['city'].str.contains('NY-NJ') == False]['monthly_revenue']
alpha = 0.05
results = st.ttest_ind(avg_rev_ny_nj, avg_rev_other_regions)
print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print('We reject the null hypothesis')
else:
    print('We cannot reject the null hypothesis')
````
