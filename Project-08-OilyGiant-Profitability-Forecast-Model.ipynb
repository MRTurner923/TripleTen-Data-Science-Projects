### Profit Calculation Function

```python
def profit_calc(predictions, target, num_wells, dev_budget):
    top_wells = sorted(range(len(predictions)), key = lambda i: predictions[i], reverse = True)[:num_wells]
    top_targets = target.iloc[top_wells]
    revenue = sum(top_targets * 4500)
    cost = dev_budget
    profit = revenue - cost
    return profit
region0_profit = profit_calc(predictions_valid0, target_valid0, num_wells, dev_budget)
region1_profit = profit_calc(predictions_valid1, target_valid1, num_wells, dev_budget)
region2_profit = profit_calc(predictions_valid2, target_valid2, num_wells, dev_budget)

print(f'The profit of region 0 is: ${region0_profit:.2f}\nThe profit of region 1 is: ${region1_profit:.2f}\nThe profit of region 2 is: ${region2_profit:.2f}')
```

### Calculating risks and profit for each region

```python
predictions_valid0 = pd.Series(predictions_valid0)
target_valid0 = target_valid0.reset_index(drop = True)
target_predict0 = pd.concat([predictions_valid0, target_valid0], axis = 1)
target_predict0.columns = ['predicted', 'actual']
target_predict0.head()
```

```python
n_samples = 1000
bootstrap_profits0 = []
state = RandomState(2356)

for i in range(n_samples):
    #Resampling data with replacement
    sample0 = target_predict0.sample(n = 500, replace=True, random_state = state)
    #Extracting predicted and actual values to plug into profit_calc function
    predictions = sample0['predicted'].reset_index(drop = True)
    targets = sample0['actual'].reset_index(drop = True)
    #Calculating profit for the region using profit_calc function and set development variables
    profit0 = profit_calc(predictions, targets, num_wells, dev_budget)
    #Appending profits list with each value for all samples taken
    bootstrap_profits0.append(round(profit0, 2))
```
